```python
def count_substrings(sequence, k=4):
    """
    Count occurrences of k-mer substrings in a given sequence.

    Args:
    sequence (str): The input string consisting of [a, c, t, g] characters.
    k (int): The length of the substrings. Default is 4.

    Returns:
    dict: A dictionary with substrings as keys and their counts as values.
    """
    # Initialize an empty dictionary to store substring counts
    counts = {}

    # Iterate over the sequence using a sliding window of size k
    for i in range(len(sequence) - k + 1):
        substring = sequence[i:i+k]

        # Increment the count for the current substring
        counts[substring] = counts.get(substring, 0) + 1

    return counts

# Example usage
sequence = "actgactgactgactg"
k = 4
result = count_substrings(sequence, k)
print(result)
```

```python
def count_unique_filenames(filename):
    try:
        # Read the file content
        with open(filename, 'r') as file:
            # Read all lines and strip whitespace
            filenames = [line.strip() for line in file if line.strip()]

        # Use a set to find unique filenames
        unique_filenames = set(filenames)

        # Print the count of unique filenames
        print(f"Number of unique filenames: {len(unique_filenames)}")

        # Optionally, print the unique filenames
        print("Unique filenames:")
        for name in unique_filenames:
            print(name)

    except FileNotFoundError:
        print(f"The file {filename} was not found.")

# Usage example:
filename = 'onion_dna_sequence.txt'  # Replace with your actual file path
count_unique_filenames(filename)
```

```python
import random

def generate_random_sequence(length=1600):
    """
    Generate a random sequence of specified length consisting of letters A, C, T, and G.

    Args:
    length (int): The length of the sequence to generate.

    Returns:
    str: A random sequence of A, C, T, and G.
    """
    characters = ['A', 'C', 'T', 'G']
    random_sequence = ''.join(random.choices(characters, k=length))
    return random_sequence

def Kmers_funct(seq, size):
    """
    Extract all k-mer substrings of a given size from a sequence.

    Args:
    seq (str): The input sequence (random in this case).
    size (int): The size of k-mers.

    Returns:
    list: A list of all k-mers in the sequence.
    """
    return [seq[x:x+size].lower() for x in range(len(seq) - size + 1)]

# Main execution
k = 5
total_substrings = 4 ** k
generation_count = 0
count = 0
next_threshold = 10  # The first threshold to notify

while count != total_substrings:
    random_sequence = generate_random_sequence()
    generation_count += 1
    k_mers = set(Kmers_funct(random_sequence, k))  # Use the optimized Kmers_funct
    count = len(k_mers)

    # Notification logic
    percentage_complete = (count / total_substrings) * 100

    # Check if we've passed the next threshold and notify
    if percentage_complete >= next_threshold:
        print(f"Passed {next_threshold}% of total substrings. Generation count: {generation_count}")
        next_threshold += 10  # Increment the threshold for the next notification

print(f"Number of random sequence generations: {generation_count}")
print(f"Random sequence: {random_sequence}")

```
